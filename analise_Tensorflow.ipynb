```python
import tensorflow as tf
from sklearn.model_selection import train_test_split
import pandas as pd
from datetime import datetime
import numpy as np
import time

import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

#Modules for ML
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.model_selection import train_test_split
from tensorflow import keras
from sklearn.metrics import classification_report,confusion_matrix
```

```python
data = pd.read_csv('OV2.csv', on_bad_lines='skip', sep=';', )
data = pd.DataFrame(data)
data.info()
```

```python
data = data.drop(data.columns[[9, 10]], axis=1)
data['Date'] = pd.to_datetime(data['Date'])
data['Date'] = pd.to_datetime(data['Date'])
data = data.dropna()
data['Hour'] = data['Date'].dt.hour
data['Minute'] = data['Date'].dt.minute
data['Weekday'] = data['Date'].dt.weekday
data['Moving_streams'] = data['Streams'].rolling(10, min_periods=1).mean()

data.info()
```

```python
X = data[['Hour','Minute','Weekday', 'Holidays']]
y = data[['Moving_streams']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```

```python
from tensorflow.keras.layers import Dense, LSTM

model = tf.keras.models.Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(4, 1)))
model.add(LSTM(units=50))
model.add(Dense(units=1))
model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(X_train, y_train, epochs=70)
```

```python
model.summary()
```

```python
model = tf.keras.models.Sequential([
  tf.keras.layers.Dense(64, activation='relu', input_shape=(4,)),
  tf.keras.layers.Dense(32, activation='relu'),
  tf.keras.layers.Dense(1)
])

model.compile(optimizer='adam', loss='mse')

model.fit(X_train, y_train, epochs=100)
```

```python
def predict(data):
    pred = model.predict(data).flatten()
    pred = pred.tolist()
    print(pred)
    print(y_test)
    return pred

y_pred_test = predict(X_test)

df = pd.DataFrame(y_pred_test)
df['should'] = y_test
df.columns = ['is','should']
display(df)

model.evaluate(X_test, y_test)
```
